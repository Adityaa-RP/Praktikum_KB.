# Import Library

```python
import pandas as pd
```

# Membaca data dari file CSV ke dalam DataFrame

```python
df = pd.read_csv("data.csv")
```

# Menampilkan rata-rata, standar deviasi, nilai minimum, nilai maksimum, Q1, Q2, dan Q3 dari attribute bertipe angka.

```python
# Memilih semua kolom dengan tipe data numerik/angka
atribut_angka = df.select_dtypes(include=['number'])

# Menghitung statistik deskriptif menggunakan describe()
stats = atribut_angka.describe()

print("Rata-rata:")
print(stats.loc['mean'])

print("\n")

print("Standar Deviasi:")
print(stats.loc['std'])

print("\n")

print("Nilai Minimum:")
print(stats.loc['min'])

print("\n")

print("Nilai Maksimum:")
print(stats.loc['max'])

print("\n")

print("Q1:")
print(atribut_angka.quantile(0.25))

print("\n")

print("Q2(Median):")
print(atribut_angka.quantile(0.50))

print("\n")

print("Q3:")
print(atribut_angka.quantile(0.75))
```

# Menampilkan informasi jumlah record, jumlah attribute, nama attribute, jumlah attribute dengan tipe angka. Masukkan informasi tersebut dalam dictionary dengan key sesuai nama informasinya.

```python
# Menghitung jumlah baris (record) dalam dataset
jumlah_record = len(df)

# Menghitung jumlah atribut (kolom) dalam dataset
jumlah_attribute = len(df.columns)

# Mengambil daftar nama atribut (kolom) dalam dataset
nama_attribute = df.columns.tolist()

# Menghitung jumlah atribut yang memiliki tipe data numerik/angka
jumlah_attribute_angka = len(df.select_dtypes(include='number').columns)

# Membuat dictionary 'informasi_dataset'
informasi_dataset = {
    "Jumlah Record          ": jumlah_record,
    "Jumlah Attribute       ": jumlah_attribute,
    "Nama Attribute         ":  nama_attribute,
    "Jumlah Attribute Angka ": jumlah_attribute_angka
}

# Mencetak informasi dataset
for key, value in informasi_dataset.items():
    print(f"{key}: {value}")
```

# Menampilkan jumlah nilai dari semua attribute yang bernilai object

```python
# Mengidentifikasi kolom dengan tipe data objek (string)
kolom_objek = df.select_dtypes(include=['object'])

# Menginisialisasi dictionary untuk menyimpan hasil perhitungan
count_results = {}

# Menggunakan loop untuk menghitung jumlah nilai dalam setiap kolom objek
for kolom in kolom_objek:
    count_results[kolom] = df[kolom].value_counts()

# Menampilkan hasil perhitungan
for kolom, hasil in count_results.items():
    print("=======================================")
    print(f"Jumlah nilai dalam {kolom}:")
    print("=======================================")
    print(hasil, "\n")
```
