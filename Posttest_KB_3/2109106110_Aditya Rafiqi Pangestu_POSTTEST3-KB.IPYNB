# Import Library

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

# Membaca data dari file CSV ke dalam DataFrame

```python
df = pd.read_csv("data.csv")
```

# Melihat informasi dari dataset

```python
df.info()
```

# Membuat Variabel Platform dan groupby

```python
Platform = df["diagnosis"].value_counts()
```

```python
df.mean = df.groupby('radius_mean').mean()
df.mean
```

```python
x = df.mean.index
y = df.mean.values
```

# 1.Grafik Batang

**Penjelasan:**
Dari grafik dibawah dapat disimpulkan diagnosis dengan M (HIJAU) yang tertinggi 17.5. Sedangkan, diagnosis B (MERAH) yang terendah 12.5.

```python
x = df['diagnosis']
y = df['radius_mean']
colors = ['green', 'red', 'yellow']
sns.barplot(
    x=x,
    y=y,
    palette=colors
)
plt.title('Grafik Diagnosisis')
plt.show()
```

# 2.Pie Chart/Diagram Lingkaran

**Penjelasan:**
Dari grafik dibawah dapat disimpulkan banyak terdiagnosis B 62.74% . Sedangkan, yang terdiagnosis M 37.26%.

```python
# Membuat pie chart(diagram lingkaran)
plt.pie(Platform, labels=None, startangle=90, autopct='%.2f%%')

# Mengambil label indeks (nama platform) dan hasil persen dari Pie chart
labels = [f'{index}: {percentage:.2f}%' for index, percentage in zip(Platform.index, (Platform / Platform.sum()) * 100)]

# Menambahkan legend dengan label yang berisi indeks dan hasil persen
# Kenapa Pakai legend? Karena label nya tertumpuk sehingga tidak terlihat tulisannya
plt.legend(labels, loc="best")
plt.legend(labels, bbox_to_anchor=(1.2, 1.1))

# Menambahkan judul (title) pada diagram
plt.title("Diagnosis")

plt.show()
```

# 3.  Histogram

**Penjelasan:**

Histogram tersebut menunjukkan frekuensi radius dari data. Dari rentan score 0 hingga 10 terjadi berbagai kenaikan dan penurunan, tapi setelah mengalami penurunan pasti terjadi kenaikan yang lebih tinggi dari kenaikan sebelumnya. Di rentan 10-15 pola serupa terjadi, tetapi setelah mendekati 15 kenaikan yang dialami setelah mengalami penurunan itu tidak lebih tinggi dari kenaikan sebelumnya. Dan di rentan 20-25 terjadi penurunan frekuensi kemunculan yang drastis.

```python
# Membuat histogram dari kolom "score" dalam DataFrame df
sns.histplot(df["radius_mean"], bins=50, color="red")
plt.xlabel("Radius")
plt.ylabel("Frekuensi Radius")

# Menambahkan judul (title) pada histogram
plt.title("Histogram Radius")

plt.show()
```

# 4.HeatMap

**Penjelasan:**
Heatmap Menyeluruh

```python
# Menghitung matriks korelasi antara kolom-kolom numerik dalam DataFrame
df.data_corr = df.corr()

# Membuat heatmap
sns.heatmap(df.data_corr, cmap="coolwarm", annot=True)

# Menambahkan judul (title) pada heatmap
plt.title("Korelasi Antar Kolom Bertipe Data Numerik")
plt.show()
```

**Penjelasan:**
Heatmap Texture dan Area Mean saja

```python
atribut_numerik = ['texture_mean','area_mean']

correlation_matrix = df[atribut_numerik].corr()

plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Heatmap Korelasi Texture dan Area')
plt.show()
```

# 5.Grafik BoxPlot

**Penjelasan:**
Pada Grafik Box Plot dibawah menunjukan 2500  area tertinggi dan 200 area terendahnya

```python
palette = "Set2"
plt.figure(figsize=(10,6))
sns.boxplot(
    data   = df,
    x      = 'texture_mean',
    y      = 'area_mean',
    palette=palette
)

plt.xlabel('Texture')
plt.ylabel('Area')
plt.title('Grafik Tingkat Texture dan juga Area')
plt.show()
```
